```python
import requests
macbeth = requests.get('https://www.gutenberg.org/files/1533/1533-0.txt').text

# Split the transcript into words
words_list = macbeth.split()

#print (words_list)
```

```python
# Create a dictionary
macbeth_dict = {}
duplicate_list = []
```

```python
# Update word counts - dict.update{ } allows you to add new key pairs
#if it's not in the list, append it. if it is, search through the list and then increment pair value 
def dictionary_generater(words_list):
    for word in words_list:
        if word not in duplicate_list:
            macbeth_dict.update({word: 1})
            duplicate_list.append(word)
        else:
            for other_word in duplicate_list:
                if other_word == word:
                    macbeth_dict[word] += 1
```
